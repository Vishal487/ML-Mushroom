```
# required libraries

import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics
```

### Data Preprocessing

```
# load training data

train_csv = pd.read_csv('drive/My Drive/train_mush.csv')
train_csv.head()
```

```
train_csv.shape
```

```
# split into X and y

X = train_csv.drop('class', axis=1)
y = train_csv['class']

X.shape, y.shape
```

```
X.head()
```

```
# check for missing values

X.isna().sum()
```

```
# check about dtypes of each column

X.info()
```

```
# converting into numerical value i.e. cat => num

X = pd.get_dummies(X)
X.head()
```

```
# see how many different classes are there?

y.unique()
```

```
y
```

```
# convert it to numerical also

y = y.astype('category').cat.codes
```

```
y             # note: e -> 0     and    p -> 1
```

**We're ready with X and y**

### Create Classification model

```
# let's split into train/val set

np.random.seed(42)

X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2)
X_train.shape, y_train.shape, X_val.shape, y_val.shape
```

```
# creating a classifier  (ensemble classifier works best here)

# instantiate the clf
clf = RandomForestClassifier(n_estimators=100)

# fit into traning set
clf.fit(X_train, y_train)
```

```
# check accuracy on validation set

clf.score(X_val, y_val)
```

```
# make prediction on validation set

y_pred = clf.predict(X_val)

# check accuracy 
metrics.accuracy_score(y_pred, y_val)
```

**now training on full training set**

```
# instantiate new classifier
new_clf = RandomForestClassifier(n_estimators=100)

# fit into FULL TRAINING DATASET
new_clf.fit(X,y)
```

### Preparing for Submission
make predction on test dataset

```
# load test set
test = pd.read_csv('drive/My Drive/Q1_Mushroom_test.csv')
test.head()
```

```
test.info()
```

```
# making test set same as train set

test_X = pd.get_dummies(test)
test_X.head()
```

```
set(X.columns) - set(test_X.columns)
```

```
set(test_X.columns) - set(X.columns)
```

Okay so we don't have **cap-shape_c** column in test set. It also mean that we don't have any sample with cap-shape = "c". So we can add a column with name **cap-shape_c** with all entries `0 (zero)`

```
# adding column named "cap-shape_c" as a second (index 1) column

test_X.insert(1, 'cap-shape_c', [0]*len(test_X))
test_X.head()
```

```
set(X.columns) - set(test_X.columns)    # ok so we got a new column
```

```
# now make prediction on test set

prediction = new_clf.predict(test_X) 
```

```
prediction.shape
```

```
# replace 0/1 with e/p

prediction_class = []
labels = ['e', 'p']
for i in prediction:
    prediction_class.append(labels[i])
```

```
# make submission dataframe

submission_mushroom = pd.DataFrame()
submission_mushroom['class'] = prediction_class
submission_mushroom.head()
```

```
submission_mushroom.shape
```

```
# save our "submission_mushroom" DF as csv file

submission_mushroom.to_csv('drive/My Drive/submission_mushroom.csv', index=False)
```
